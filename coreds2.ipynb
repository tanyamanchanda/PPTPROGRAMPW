```python
import statistics
import scipy.stats

# Scenario 1: Calculate the mean sales for each region
region_a_sales = [10, 15, 12, 8, 14]
region_b_sales = [18, 20, 16, 22, 25]

mean_region_a_sales = statistics.mean(region_a_sales)
mean_region_b_sales = statistics.mean(region_b_sales)

print("Mean sales for Region A:", mean_region_a_sales)
print("Mean sales for Region B:", mean_region_b_sales)
print()

# Scenario 2: Calculate the mode of the survey responses
survey_responses = [4, 5, 2, 3, 5, 4, 3, 2, 4, 5]
mode_survey_responses = statistics.mode(survey_responses)

print("Mode of survey responses:", mode_survey_responses)
print()

# Scenario 3: Calculate the median salary for each department
department_a_salaries = [5000, 6000, 5500, 7000]
department_b_salaries = [4500, 5500, 5800, 6000, 5200]

median_department_a_salaries = statistics.median(department_a_salaries)
median_department_b_salaries = statistics.median(department_b_salaries)

print("Median salary for Department A:", median_department_a_salaries)
print("Median salary for Department B:", median_department_b_salaries)
print()

# Scenario 4: Calculate the range of the stock prices
stock_prices = [25.5, 24.8, 26.1, 25.3, 24.9]
stock_price_range = max(stock_prices) - min(stock_prices)

print("Range of stock prices:", stock_price_range)
print()

# Scenario 5: Perform a t-test to determine if there is a significant difference in the mean scores between the two groups
group_a_scores = [85, 90, 92, 88, 91]
group_b_scores = [82, 88, 90, 86, 87]

t_statistic, p_value = scipy.stats.ttest_ind(group_a_scores, group_b_scores)

print("T-statistic:", t_statistic)
print("P-value:", p_value)
print()

# Scenario 6: Calculate the correlation coefficient between advertising expenditure and sales
advertising_expenditure = [10, 15, 12, 8, 14]
sales = [25, 30, 28, 20, 26]

correlation_coefficient = scipy.stats.pearsonr(advertising_expenditure, sales)[0]

print("Correlation coefficient:", correlation_coefficient)
```

```python
import statistics
import scipy.stats
import numpy as np
import statsmodels.api as sm

# Scenario 7: Calculate the standard deviation of the heights
heights = [160, 170, 165, 155, 175, 180, 170]
std_dev_heights = statistics.stdev(heights)

print("Standard deviation of heights:", std_dev_heights)
print()

# Scenario 8: Perform a linear regression analysis to predict job satisfaction based on employee tenure
employee_tenure = [2, 3, 5, 4, 6, 2, 4]
job_satisfaction = [7, 8, 6, 9, 5, 7, 6]

X = sm.add_constant(employee_tenure)
model = sm.OLS(job_satisfaction, X)
results = model.fit()

print("Linear regression results:")
print(results.summary())
print()

# Scenario 9: Perform an analysis of variance (ANOVA) to determine if there is a significant difference in the mean recovery times between the two medications
medication_a_recovery_times = [10, 12, 14, 11, 13]
medication_b_recovery_times = [15, 17, 16, 14, 18]

f_statistic, p_value = scipy.stats.f_oneway(medication_a_recovery_times, medication_b_recovery_times)

print("ANOVA results:")
print("F-statistic:", f_statistic)
print("P-value:", p_value)
print()

# Scenario 10: Calculate the 75th percentile of the feedback ratings
feedback_ratings = [8, 9, 7, 6, 8, 10, 9, 8, 7, 8]
percentile_75 = np.percentile(feedback_ratings, 75)

print("75th percentile of feedback ratings:", percentile_75)
print()

# Scenario 11: Perform a hypothesis test to determine if the mean weight differs significantly from 10 grams
weights = [10.2, 9.8, 10.0, 10.5, 10.3, 10.1]

t_statistic, p_value = scipy.stats.ttest_1samp(weights, 10)

print("Hypothesis test results:")
print("T-statistic:", t_statistic)
print("P-value:", p_value)
print()

# Scenario 12: Perform a chi-square test to determine if there is a significant difference in the click-through rates between the two designs
design_a_clicks = [100, 120, 110, 90, 95]
design_b_clicks = [80, 85, 90, 95, 100]

chi2_statistic, p_value, _, _ = scipy.stats.chi2_contingency([design_a_clicks, design_b_clicks])

print("Chi-square test results:")
print("Chi2-statistic:", chi2_statistic)
print("P-value:", p_value)
print()

# Scenario 13: Calculate the 95% confidence interval for the population mean satisfaction score
satisfaction_scores = [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]

confidence_interval = scipy.stats.t.interval(0.95, len(satisfaction_scores)-1, loc=np.mean(satisfaction_scores), scale=scipy.stats.sem(satisfaction_scores))

print("95% confidence interval for population mean satisfaction score:", confidence_interval)
print()

# Scenario 14: Perform a simple linear regression to predict performance based on temperature
temperature = [20, 22, 23, 19, 21]
performance = [8, 7, 9, 6, 8]

X = sm.add_constant(temperature)
model = sm.OLS(performance, X)
results = model.fit()

print("Linear regression results:")
print(results.summary())
print()

# Scenario 15: Perform a Mann-Whitney U test to determine if there is a significant difference in the median preferences between the two groups
group_a_preferences = [4, 3, 5, 2, 4]
group_b_preferences = [3, 2, 4, 3, 3]

U_statistic, p_value = scipy.stats.mannwhitneyu(group_a_preferences, group_b_preferences)

print("Mann-Whitney U test results:")
print("U-statistic:", U_statistic)
print("P-value:", p_value)
print()

# Scenario 16: Calculate the interquartile range (IQR) of the ages
ages = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
iqr_ages = np.percentile(ages, 75) - np.percentile(ages, 25)

print("Interquartile range (IQR) of ages:", iqr_ages)
print()

# Scenario 17: Perform a Kruskal-Wallis test to determine if there is a significant difference in the median accuracy scores between the algorithms
algorithm_a_scores = [0.85, 0.80, 0.82, 0.87, 0.83]
algorithm_b_scores = [0.78, 0.82, 0.84, 0.80, 0.79]
algorithm_c_scores = [0.90, 0.88, 0.89, 0.86, 0.87]

H_statistic, p_value = scipy.stats.kruskal(algorithm_a_scores, algorithm_b_scores, algorithm_c_scores)

print("Kruskal-Wallis test results:")
print("H-statistic:", H_statistic)
print("P-value:", p_value)
print()

# Scenario 18: Perform a simple linear regression to predict sales based on price
price = [10, 15, 12, 8, 14]
sales = [100, 80, 90, 110, 95]

X = sm.add_constant(price)
model = sm.OLS(sales, X)
results = model.fit()

print("Linear regression results:")
print(results.summary())
print()

# Scenario 19: Calculate the standard error of the mean satisfaction score
satisfaction_scores = [7, 8, 9, 6, 8, 7, 9, 7, 8, 7]
standard_error = scipy.stats.sem(satisfaction_scores)

print("Standard error of the mean satisfaction score:", standard_error)
print()

# Scenario 20: Perform a multiple regression analysis to predict sales based on advertising expenditure
advertising_expenditure = [10, 15, 12, 8, 14]
sales = [25, 30, 28, 20, 26]

X = sm.add_constant(advertising_expenditure)
model = sm.OLS(sales, X)
results = model.fit()

print("Multiple regression results:")
print(results.summary())
print()
```
